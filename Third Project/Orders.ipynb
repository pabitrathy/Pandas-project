```python
import pandas as pd
```

```python
df = pd.read_csv("Orders.csv")
```

```python
df
```

```python
df["Order_Date"] = pd.to_datetime(df["Order_Date"])
```

```python
df.dtypes
```

```python
df["Order_Month"] = df["Order_Date"].dt.month_name()
```

```python
df
```

```python
df["Total_Amount"] = df["Quantity"]*df["Unit_Price"]
```

```python
df
```

```python
total_sales = df.groupby("Category")["Quantity"].sum()
```
